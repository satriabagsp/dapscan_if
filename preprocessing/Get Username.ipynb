```python
import pandas as pd
import random
import itertools
import numpy as np
```

```python
df_daftar_nama = pd.read_excel('../Data/Nama untuk sistem (1).xlsx')

# Clean teks
def clean_teks(teks):
    teks = ' '.join(teks.split())
    return teks

df_daftar_nama['Nama'] = df_daftar_nama['Nama'].apply(clean_teks)

df_daftar_nama = df_daftar_nama[['Nama']].drop_duplicates().reset_index(drop=True)

df_daftar_nama = df_daftar_nama.rename(columns={'Nama':'name'})

df_daftar_nama
```

```python
nama = 'Indah Budiati SST, M.Si'

# CLEAN NAMA TO GET USERNAME

def get_username(nama):
    list_nama = nama.split(' ')

    username = []
    for nama in list_nama:
        if '.' in nama or ',' in nama:
            pass
        elif nama.isupper():
            pass
        else:
            username.append(nama.lower())

    username = '_'.join(username)

    return username

df_daftar_nama['username'] = df_daftar_nama['name'].apply(get_username)

df_daftar_nama.to_csv('../Data/list_username.csv', sep = ';', index=False)
df_daftar_nama
```
